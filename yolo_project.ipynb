```python
public_data='darknet/dataset/public/'
private_data='darknet/dataset/private/'
```

```python
from subprocess import Popen, PIPE
#./darknet detect cfg/yolov3.cfg yolov3.weights dataset/public/4380220266.jpg

def detect_object(image):
    p = Popen(['./darknet', 'detect','cfg/yolov3.cfg', 'yolov3.weights', image],
    cwd = '/home/aindani/darknet', stdout = PIPE, stderr = PIPE)
    stdout, stderr = p.communicate()
    output = stdout.decode("utf-8")
    error = stderr.decode("utf-8")
    return output,stderr
```

```python
#single file testing the function
out,error = detect_object('dataset/public/4381895254.jpg')
print (out)
```

```python
import glob
public_images=glob.glob(public_data+'*.jpg')
'/'.join(public_images[0].split('/')[1:])
```

```python
public_ans_dict={}
for file in public_images:
    cur_file='/'.join(file.split('/')[1:])
    out,error = detect_object(cur_file)
    if out.strip() == '':
        print("Error : " + str(error))
        break
    print(out.split('\n')[0])
    for line in out.split('\n')[1:]:
        if '%' in line:
            print(line)
            key=line.split(':')[0].strip()
            if key in public_ans_dict.keys():
                public_ans_dict[key] = public_ans_dict[key] + 1
            else:
                public_ans_dict[key] =1
```

```python
import glob
private_images=glob.glob(private_data+'*.jpg')
```

```python
private_ans_dict={}
for file in private_images:
    cur_file='/'.join(file.split('/')[1:])
    out,error = detect_object(cur_file)
    if out.strip() == '':
        print("Error : " + str(error))
        break
    print(out.split('\n')[0])
    for line in out.split('\n')[1:]:
        if '%' in line:
            print(line)
            key=line.split(':')[0].strip()
            if key in private_ans_dict.keys():
                private_ans_dict[key] = private_ans_dict[key] + 1
            else:
                private_ans_dict[key] =1
    
```
